## Global Keyword
- The global keyword is used inside a function to indicate that it refers to global variable not local variable

```python
x=10
def update():
    x=x+6
    return x
print(x)
update()
```

```python
x=10
def update():
    global x
    x=x+6
    return x
print(x)
update()
```

## Globals()
- globals() is a built-in function that returns a dictionary representing the current global symbol table.
- It allows you to access and modify global variables programmatically.

```python
x=10
def update():
    globals()['x']=x+15 #acess & modify the global variable using dictionary
    return x
print(x)
update()
```

## Passing List to a Function

```python
# Count of Even & Odd numbers in a list using Functions
def count(lst):
    even=0
    odd=0
    for i in lst:
        if i%2==0:
            even+= 1
        else:
            odd+= 1
    print('EvenCount:',even)
    print('OddCount:',odd)
nums=[1,7,8,17,54,18,2,6,10]
count(nums)
```

```python
# Fibonnaci Sequence
def fib(n):
    a,b=0,1
    if n<=1:
        return a
    else:
        print(a)
        print(b)
        for i in range(0,n):
            c=a+b
            a=b
            b=c
            print(c)
fib(10)
```

```python
# Factorial
def fact(n):
    res=1
    for i in range(1,n+1):
        res=res*i
    return res
fact(5)
```

```python
# Factorial
def fact(n):
    if n==0:
        return 1
    else:
        return n*fact(n-1)
fact(5)
```

## Anonymous or Lambda Function
- a function without a name called as anonymous or lambda function

```python
# Normal Function
def square(n):
    return n*n
square(6)
```

```python
# Lambda Function
s=lambda n:n*n #argument,operation
s(6)
```

```python
add=lambda a,b:a+b
add(18,42)
```

## filter()
- filter() is used to filter elements from an iterable based on a function that returns True or False.

```python
nums=[1,5,7,6,8,6,3,8,5,9,0]
```

```python
def is_even(n):
    return n%2==0
list(filter(is_even,nums))
```

```python
list(filter(lambda n:n%2==0,nums))
```

```python
def is_odd(n):
    return n%2!=0
list(filter(is_odd,nums))
```

```python
list(filter(lambda n:n%2!=0,nums))
```

## map()
- map() applies a function to all items in an iterable and returns a map object, which can be converted into a list or another iterable type.

```python
nums=[1,2,3,4,5]
def add2(n):
    return n+2
list(map(add2,nums))
```

```python
list(map(lambda n:n*2,nums)) #square of nums
```

## reduce()
- reduce() is part of the functools module and is used to apply a function cumulatively to the elements of an iterable.

```python
# we have to import reduce from functoolsn
from functools import reduce
def add_all(a,b):
    return a+b

reduce(add_all,nums)
```

```python
reduce(lambda a,b:a+b,nums)
```

```python
reduce(lambda a,b:a*b,nums)
```

## Decorator
- A decorator in Python is a design pattern that allows you to modify the behavior of a function or method without changing its actual code. It wraps a function inside another function to extend its functionality.

```python
def decorator_function(original_function):
    def wrapper_function():
        print("Wrapper executed before", original_function.__name__)
        return original_function()
    return wrapper_function

@decorator_function
def display():
    print("Display function executed")
display()
```

```python
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")
say_hello()
```
